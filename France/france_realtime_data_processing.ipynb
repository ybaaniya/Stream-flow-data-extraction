```python
import os
import pandas as pd
```

```python
# Define the master directory containing the subfolders
master_directory = 'folder containing downloaded realtime data'
output_directory = 'folder to sotre processed realtime data'
```

```python
# Iterate over each subfolder in the master directory
for subfolder in os.listdir(master_directory):
    subfolder_path = os.path.join(master_directory, subfolder)
    
    # Skip if it's not a directory
    if not os.path.isdir(subfolder_path):
        continue
    
    print(f'Processing subfolder: {subfolder}')
    
    # Initialize an empty list to hold the dataframes
    dataframes = []
    
    # Iterate over the files in the subfolder
    for filename in os.listdir(subfolder_path):
        if filename.endswith(".csv"):
            try:
                # Extract dates and station number from the filename
                parts = filename.split('_')
                station_number = parts[0]
                start_date = parts[1]
                end_date = parts[3].replace('.csv', '')

                # Read the CSV file
                filepath = os.path.join(subfolder_path, filename)
                df = pd.read_csv(filepath)
                
                # Add start_date and end_date columns
                df['start_date'] = pd.to_datetime(start_date, format='%d-%m-%Y')
                df['end_date'] = pd.to_datetime(end_date, format='%d-%m-%Y')
                
                # Add the dataframe to the list
                dataframes.append(df)
            except Exception as e:
                print(f'Error processing file {filename}: {e}')
    
    if dataframes:
        # Concatenate all dataframes
        combined_df = pd.concat(dataframes)

        # Convert the 't' column to datetime for proper sorting
        combined_df['t'] = pd.to_datetime(combined_df['t'])

        # Sort the combined dataframe by 't' column
        combined_df = combined_df.sort_values(by='t')

        # Drop the temporary date columns
        combined_df = combined_df.drop(columns=['start_date', 'end_date'])

        # Save the combined dataframe to a new CSV file in the output directory
        output_filepath = os.path.join(output_directory, f'{station_number}.csv')
        combined_df.to_csv(output_filepath, index=False)
        
        print(f'Successfully processed and saved: {station_number}')
    else:
        print(f'No CSV files found in subfolder: {subfolder}')
```
